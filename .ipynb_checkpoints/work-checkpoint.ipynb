```python
df2 = pd.io.parsers.read_csv('2001_Urban_Households_Lighting.csv')
df2 = df2.append(pd.io.parsers.read_csv('2001_Rural_Households_Lighting.csv'))
df2 = df2.append(pd.io.parsers.read_csv('2011_Rural_Households_Lighting.csv'))
df2 = df2.append(pd.io.parsers.read_csv('2011_Urban_Households_Lighting.csv'))
```

```python
def f(s):
    return s[:len(s)/2]

year = 2001
a = df2[(df2.Type == 'Rural') & (df2.Year == year)].combineAdd(df2[(df2.Type == 'Urban') & (df2.Year == year)])
a['Year'] = year
a['Type'] = 'Total'
a['State'] = a['State'].apply(f)
df2 = df2.append(a)

year = 2011
a = df2[(df2.Type == 'Rural') & (df2.Year == year)].combineAdd(df2[(df2.Type == 'Urban') & (df2.Year == year)])
a['Year'] = year
a['Type'] = 'Total'
a['State'] = a['State'].apply(f)
df2 = df2.append(a)
```

```python
df2['Elec_Perc'] = df2.Electricity/df2.Total
df2['Kerosene_Perc'] = df2.Kerosene/df2.Total
df2['Solar_Perc'] = df2.Solar/df2.Total
df2['No_lighting_Perc'] = df2['No_lighting']/df2.Total
df2 = df2.sort('State')
df2 = df2[(df2.State != 'Lakshadweep') & (df2.State != 'Andaman & Nicobar Islands')  & (df2.State != 'Chandigarh')  & (df2.State != 'Dadra & Nagar Haveli') & (df2.State != 'Daman & Diu') & (df2.State != 'India')]
```

```python
states = df2[(df2.Year == 2011) & (df2.Type == 'Total')].State.values
elec_perc = df2[(df2.Year == 2011) & (df2.Type == 'Total') & (df2.State != 'Arunachal Pradesh')].Elec_Perc.values
```

```python
import statsmodels.api as sm

Y = df[df.State != 'Arunachal Pradesh']['2011-12'].values
Y = Y/100

no_light_perc = df2[(df2.Year == 2011) & (df2.Type == 'Total') & (df2.State != 'Arunachal Pradesh')].No_lighting_Perc.values
solar_perc = df2[(df2.Year == 2011) & (df2.Type == 'Total') & (df2.State != 'Arunachal Pradesh')].Solar_Perc.values
ker_perc = df2[(df2.Year == 2011) & (df2.Type == 'Total') & (df2.State != 'Arunachal Pradesh')].Kerosene_Perc.values
#print np.corrcoef(ker_perc, td_losses)[0][-1]
#print np.corrcoef(no_light_perc, td_losses)[0][-1]
#r = np.corrcoef(elec_perc, td_losses)[0][-1]
#print 'r:', r, 'r^2:', r**2
#print np.corrcoef(solar_perc, td_losses)[0][-1]

#plt.scatter(ker_perc, Y)
#plt.title('Percentage of households that use kerosene for lighting against TD losses')
#plt.show()

a = no_light_perc
b = Y
c = elec_perc

plt.scatter(a, b)
plt.title('Percentage of households that do not have lighting in a state against\nT&D losses of the state')
p = np.polyfit(a,b,1)
plt.plot(a,p[0]*a+p[1])
plt.show()
r = np.corrcoef(a, b)[0][-1]
print 'r:', r, 'r^2:', r**2

plt.scatter(c, b)
plt.title('Percentage of households that use electricity for lighting in a state against\nT&D losses of the state')
p = np.polyfit(c,b,1)
plt.plot(c,p[0]*c+p[1])
plt.show()
r = np.corrcoef(c, b)[0][-1]
print 'r:', r, 'r^2:', r**2

plt.scatter(solar_perc, b)
plt.title('Percentage of households that use solar for lighting in a state against\nT&D losses of the state')
p = np.polyfit(solar_perc,b,1)
plt.plot(solar_perc,p[0]*solar_perc+p[1])

plt.show()

r = np.corrcoef(solar_perc, b)[0][-1]
print 'r:', r, 'r^2:', r**2

#plt.scatter(solar_perc, Y)
#plt.title('Percentage of households that use Solar for lighting against TD losses')
#plt.show()
```

```python
#X = pd.DataFrame()
a = df2[(df2.Year == 2011) & (df2.Type == 'Total')]
X = a[['Elec_Perc']]
#X['log_elec'] = np.log1p(a['Elec_Perc'])
X['no_light'] = a['No_lighting_Perc']
X['solar'] = a['Solar_Perc']
#X['elec_solar_int'] = a['Elec_Perc']*a['Solar_Perc']

X_mod = X.drop(X.index[1])
Y_mod = Y

#X = sm.add_constant(X_mod)
X_mod = sm.add_constant(X_mod)
est = sm.OLS(Y_mod, X_mod).fit()
#est = sm.OLS(Y, X).fit()
est.summary()
```

```python
estimates = est.params.values
input_vals = X_mod.values
predicted_vals =  np.dot(input_vals,estimates)
state_names = np.concatenate((states[0:1], states[2:]), axis=0)
df4 = pd.DataFrame()
df4['State'] = state_names
df4['Actual'] = Y_mod
df4['Predicted'] = predicted_vals
df4['Difference'] = (df4['Actual'] - df4['Predicted'])/df4['Actual']
df4['Performance'] = df4['Difference'].apply(lambda x: 'UnderPerforming' if x > 0.2 else 'OverPerforming' if x < -0.2 else 'Normal')
df4
```
