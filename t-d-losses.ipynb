### Looking at the T&D losses across different states

Loading data for T&D losses. This was obtained from http://data.gov.in/catalog/transmission-and-distribution-losses-percentage-availability-state-power-utilities-spus#web_catalog_tabs_block_10.

Only data till 2011-12 was used since data beyond that planned (instead of reported) data.

```python
df = pd.read_csv('data/losses_percentage_0913.csv').dropna()
df = df.drop(df.columns[0], axis=1)
df.columns = ['State'] + ['09-10', '10-11', '11-12', '12-13', '13-14']
df = df[['State', '09-10', '10-11', '11-12']].sort('State').reset_index().drop('index', axis=1)
df.head()
```

Pondicherry was excluded from that dataset since much related data was not available for it.

```python
df = df[(df.State != 'Pondicherry')]
```

Zipping state names and corresponding T&D losses to a single list.

```python
states = df.State.values
td_losses = df['11-12'].values
display_info = zip(states, td_losses)
```

### Setting up a new data-frame to compare T&D losses with other factors

The code below is terrible. I will clean it up when things are less busy. Also, had initially included statistics that are currently listed in unused data, but have not included them here in the interest of brevity.

Data sources:
* Data about the length of transmission lines was taken from http://mospi.nic.in/Mospi_New/upload/infra_stat_2014/electricity_publication_no1.xls
* Data about the area of each state was taken from http://en.wikipedia.org/wiki/List_of_states_and_territories_of_India_by_area
* Data about the statewise per capita consumption of electricity was taken from  http://en.wikipedia.org/wiki/Electricity_sector_in_India#Electricity_consumption

```python
a = pd.read_csv('data/per_cap_elec.csv').sort('State').dropna()
b = pd.read_csv('data/area.csv').sort('State').dropna()
states = np.unique(df.State.values)
a = a[a.State.isin(states)]
b = b[b.State.isin(states)]
area = b.Area.values
per_cap_elec = a['PerCapElec'].values/1000
```

```python
df2 = pd.io.parsers.read_csv('data/t_line_len.csv')
df2 = df2[df2.State.isin(states)]
cm_per_area = 1.*df2['2011-12'].values/area
y = df['11-12'].values
Y = pd.DataFrame()
Y['td_losses'] = y
```

```python
X = pd.DataFrame()
X['State'] = states
X['Per_cap_elec'] = per_cap_elec
X['Lines_per_area'] = cm_per_area
X['Lines_per_area_log'] = np.log(cm_per_area)
X['td_losses'] = y
```

### Plotting graphs

```python
from scipy import optimize
from scipy import interpolate

def f(x, a, b, c):
    return a*np.exp(-b*x) + c

d = {'Elec_Perc': 'Percentage of households in a state that use electricity for lighting',
     'No_lighting_Perc': 'Percentage of households in a state that do not have any avenue for lighting',
     'Solar_Perc': 'Percentage of households in a state that use solar energy for lighting',
     'Per_cap_elec': 'Per capita consumption of electricity in a state',
     'Lines_per_unit': 'Circuit Kilometres of transmission lines per unit of electricity consumer in a state',
     'Lines_per_area': 'Circuit Kilometres of transmission lines per square kilometer in a state',
     'Lines_per_area_log': 'log(Circuit Kilometres of transmission lines per square kilometer) in a state'}

#for param in ['Elec_Perc', 'No_lighting_Perc', 'Solar_Perc', 'Per_cap_elec', 'Lines_per_area_log', 'Lines_per_area']:
for param in ['Lines_per_area_log', 'Lines_per_area']:
    x = X[param].values
    y = X['td_losses'].values
    if param == 'Lines_per_area':
        a = zip(x, y)
        a.sort(key = lambda i: i[0])
        x_new, y_new = zip(*a)
        x, y = list(x_new), list(y_new)
    
    plt.scatter(x,y)
    
    if param == 'Lines_per_area':
        r = np.corrcoef(np.log(x), y)[0][-1]
        print param, 'r:', r, 'r^2:', r**2

        popt, pcov = optimize.curve_fit(f, x, y)
        x_trial = np.linspace(x[0], x[-1], 1000)
        yEXP = f(x_trial, *popt)
        plt.plot(x_trial, yEXP, alpha = 0.5)
        plt.title('%s against\nT&D losses of the state (2011)'%d[param])
    else:
        r = np.corrcoef(x,y)[0][-1]
        print param, 'r:', r, 'r^2:', r**2
        p = np.polyfit(x,y,1)
        plt.plot(x,p[0]*x+p[1], alpha = 0.5)
        plt.title('%s\nagainst T&D losses of the state (2011)'%d[param])
        xmin = int(min(x)) - 1
    
    plt.xlabel(d[param][:-10].lower())
    plt.ylabel('T&D losses as percentage of output (%)')
    plt.text(-2.2,-2,'Source: Planning Commission (http://goo.gl/23RIkk, http://goo.gl/iAXFL4)\nAuthor: @NumerateIndian | TheNumerateIndian.com')
    plt.gca().grid(True)
    plt.show()
```
